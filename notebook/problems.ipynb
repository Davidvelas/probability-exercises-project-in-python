## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO
import numpy as np
import pandas as pd
```

```python
dice1 = np.random.randint(1,7,size = 1000)
dice2 = np.random.randint(1,7,size = 1000)

result = dice1 + dice2

greater_7 = np.sum(result > 7)
evens = np.sum(result %2 == 0)

print('Number of times sum is greater then 7: ', greater_7)
print ('Number of times sum are even: ', evens)


probability1 = greater_7 / 1000
probability2 = evens / 1000 

print('Probability for greater then 7: ', probability1)
print ('Probability for even numbers: ', probability2)
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
#np.random.seed(0)
balls = ['white'] * 30 + ['red'] * 20 + ['green'] * 10

# Function to draw 5 balls from the bag
def draw_balls():
    return np.random.choice(balls, 5, replace=False)

# Simulate 1000 trials
draws = [draw_balls() for _ in range(1000)]


df = pd.DataFrame(draws, columns=['ball1', 'ball2', 'ball3', 'ball4', 'ball5'])


count_3_white_2_red = df.apply(lambda row: (row == 'white').sum() == 3 and (row == 'red').sum() == 2, axis=1).sum()
count_all_same_color = df.apply(lambda row: len(set(row)) == 1, axis=1).sum()

total_trials = len(df)
prob_3_white_2_red = count_3_white_2_red / total_trials
prob_all_same_color = count_all_same_color / total_trials

print(f"Probability of 3 white and 2 red: {round(prob_3_white_2_red * 100, 2)}%  ")
print(f"Probability of all balls being the same color: {round(prob_all_same_color * 100, 2)}%")
print(balls)
```
